```python
import pandas as pd

INPUT_CSV = "Kepler.csv"
OUTPUT_CSV = "Kepler.csv"

SELECTED_LINES = [0, 6, 8, 10, 12]

df = pd.read_csv(INPUT_CSV)

selected = df.loc[SELECTED_LINES].copy()
selected["original_row"] = selected.index
selected = selected[["original_row", "claim", "truth"]]

selected.to_csv(OUTPUT_CSV, index=False)
print(f"Saved {len(selected)} rows to {OUTPUT_CSV}")
print("Rows kept:", SELECTED_LINES)
```
